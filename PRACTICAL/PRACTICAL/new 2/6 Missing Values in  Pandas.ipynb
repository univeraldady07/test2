```python
import pandas as pd
```

```python
df=pd.read_csv('/content/Book1.csv')
```

1. Drop the column where any of the element is missing value

```python
df.dropna(how='any',axis=0)
```

2. Drop the column where all of the element is missing value

```python
df.dropna(how='all',axis=0,inplace=True)
```

```python
df
```

3. Fill the Missing values using mean, median, Max, Min

```python
mathsmean=df['Maths'].mean()
```

```python
mathsmean
```

```python
englishmedian=df['English'].median()
```

```python
englishmedian
```

```python
marathimax=df['Marathi'].max()
marathimax
```

```python
df['Maths']=df['Maths'].fillna(mathsmean)
```

```python
df
```

```python
df['English']=df['English'].fillna(englishmedian)
df
```

```python
df['Marathi']=df['Marathi'].fillna(marathimax)
```

```python
df
```

```python
Hindimin=df['Hindi'].min()
Hindimin
```

```python
df['Hindi']=df['Hindi'].fillna(Hindimin)
df
```
